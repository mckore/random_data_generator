```python
import numpy as np
```

```python
np.random.seed(2)
df = pl.DataFrame({
    f'dist_{i}': np.random.rand(10) * np.random.rand(1) for i in range(8)
}).select([pl.when(pl.col("dist_0")< 0.15).then("a").otherwise('b').alias('groups'),
pl.all()])
df
```

```python
#creating new df shape based off of aggregation
stmt = pl.all().exclude('groups')
df.select([
    stmt.sum().suffix("_sum"),
    stmt.entropy().suffix("_entropy")
    ])
```

```python
my_stmt = (pl.when(stmt.sum() >1)
        .then(stmt.entropy())
        .otherwise(stmt.skew()*stmt.entropy()*10))

df.select([my_stmt])
```

```python
(df.groupby('groups')
.agg([my_stmt,
        pl.all().first().suffix("_first")])
)
```

```python
#eager api

df.select([
    pl.concat_list(stmt).arr.eval(
        pl.when(pl.element().sum() >1)
        .then(pl.element().entropy())
        .otherwise(pl.element().skew()*pl.element().entropy()*10)
    ).flatten()
])
```

```python
#lazy execution
df.lazy().select([
    pl.concat_list(stmt).arr.eval(
        pl.when(pl.element().sum() >1)
        .then(pl.element().entropy())
        .otherwise(pl.element().skew()*pl.element().entropy()*10)
    ).flatten()
]).collect()
```
