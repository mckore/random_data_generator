```python
import csv
```

```python
#csv.reader
with open('fake_records.csv', 'r', newline = '') as file:
    reader = csv.reader(file)
    for row in reader:
        print(','.join(row))
```

```python
#csv.DictReader
with open('fake_records.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row['city'], ',', row['state'])
```
